```python
import pandas as pd
from astropy.table import Table, vstack
import numpy as np
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
PATH = Path("..")
```

```python
#Reading the files
df =  pd.read_csv(PATH / "Fornax.csv")
df1 =  pd.read_csv(PATH / "confirmedLSBs_Fornax.csv")
df2 =  pd.read_csv(PATH / "LSBs_found_23Fornax_prob08-juliana_allinf.csv")
```

```python
print("Table 1:", df.columns)
print("Table 2:", df1.columns) #names of the columns of df and df2 are the same
```

```python
# Definition for the colors
def colour(table, f1, f2, f3, f4):
    xcolour = table[f1] - table[f2]
    ycolour = table[f3] - table[f4]
    return xcolour, ycolour
```

```python
# (Colors: r - z) vs (g - r) 
cx, cy = colour(df, "g_auto", "r_auto", "r_auto", "z_auto")
cx1, cy1 = colour(df1, "G_auto", "R_auto", "R_auto", "Z_auto")
cx2, cy2 = colour(df2, "g_auto", "r_auto", "r_auto", "z_auto")
```

```python
#Plot
color_palette = sns.color_palette('Paired', 12)
with sns.axes_style('white'):
    fig, ax = plt.subplots(figsize=(14, 11))
    ax.spines["top"].set_visible(False)  
    ax.spines["right"].set_visible(False)
    plt.xlabel(r"$g - r$", fontsize=35)
    plt.ylabel(r"$r - z$", fontsize=35)
    plt.tick_params(axis='x', labelsize=35) 
    plt.tick_params(axis='y', labelsize=35)

    scatter = ax.scatter(
       cx, cy,
       s=5, 
        color="gray",
       cmap="seismic", alpha = 0.5, zorder=1.0, label="LSBG candidates")
    scatter = ax.scatter(
       cx1, cy1,
       s=50, 
        c=sns.xkcd_rgb["dark pink"],
       cmap="seismic",  edgecolor='black', alpha=0.8, zorder=4.0, label="Confirmed")
    scatter = ax.scatter(
       cx2, cy2,
       s=50, 
        c=sns.xkcd_rgb["dark pink"],
       cmap="seismic",  edgecolor='black', alpha=0.8, zorder=4.0)
    ax.set(
    xlim=[-10, 10],
    ylim=[-10, 10])
    plt.legend(fontsize="x-large")
    
```

```python
# Colors: (g - J0430) vs (J0515 - g) 
cx_g, cy_g = colour(df, "J0515_auto", "g_auto", "g_auto", "J0430_auto")
cx1_g, cy1_g = colour(df1, "F515_auto", "G_auto", "G_auto", "F430_auto")
cx2_g, cy2_g = colour(df2, "J0515_auto", "g_auto", "g_auto", "J0430_auto")
```

```python
with sns.axes_style('white'):
    fig, ax = plt.subplots(figsize=(14, 11))
    ax.spines["top"].set_visible(False)  
    ax.spines["right"].set_visible(False)
    plt.xlabel(r"$J0515 - g$", fontsize=35)
    plt.ylabel(r"$g - J0430$", fontsize=35)
    plt.tick_params(axis='x', labelsize=35) 
    plt.tick_params(axis='y', labelsize=35)

    scatter = ax.scatter(
       cx_g, cy_g,
       s=5, 
        color="gray",
       cmap="seismic", alpha = 0.5, zorder=1.0, label="LSBG candidates")
    scatter = ax.scatter(
       cx1_g, cy1_g,
       s=50, 
        c=sns.xkcd_rgb["dark pink"],
       cmap="seismic",  edgecolor='black', alpha=0.8, zorder=4.0, label="Confirmed")
    scatter = ax.scatter(
       cx2_g, cy2_g,
       s=50, 
        c=sns.xkcd_rgb["dark pink"],
       cmap="seismic",  edgecolor='black', alpha=0.8, zorder=4.0)
    ax.set(
    xlim=[-10, 10],
    ylim=[-10, 10])
    plt.legend(fontsize="x-large")
    
```
